```python
#smtp : simple mail transfer protocol
import smtplib
```

```python
#to work with passwords
import getpass
```

```python
#MIME: Multi-Purpose Internet Mail Extensions....basically it is the format of email used for exchange of data
from email.mime.text import MIMEText
```

```python
#function to send mail
def send_email():
    sender_address ='akashmallgo49@gmail.com'
    password = getpass.getpass()
    subject = 'MALL chala tarakee ki or'
    msg = '''
        Hi dosto!
        I am planning to be better than what I was yesterday.
        Hitherto I was passive but my active mode is on.
        Hope u will recieve many more of such emails from me.
        I wish to take you along as well in this learning journey
        Hope you will join!
        
        thank you!!!
        Akash Mall
    '''
    recipients = ["amitmehra10a200@gmail.com","shubham.semwal4@gmail.com"]
    #port for gmail is 587. Its different for different mail service like yahoo or hotmail
    server = smtplib.SMTP("smtp.gmail.com",587)
    #tls handshake
    server.starttls()
    server.login(sender_address,password)
    #initialise MIMETEXT
    msg = MIMEText(msg)
    
    msg['Subject'] = subject
    msg['From'] = sender_address
    msg['To'] = ",".join(recipients)
    server.sendmail(sender_address,recipients,msg.as_string())
send_email()
    
```

```python
from sklearn.datasets import make_moons
```

```python
from sklearn.cluster import DBSCAN
```

```python
x,y = make_moons(n_samples = 200,noise = 0.1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
obj = DBSCAN(eps = 0.21,min_samples = 5)
```

```python
y_pred = obj.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred)
```
