```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("./Train/Diabetes_XTrain.csv")
dfy = pd.read_csv("./Train/Diabetes_YTrain.csv")
dfxtest = pd.read_csv("./Test/Diabetes_XTest.csv")
dfy.shape
```

```python
x = dfx.values
y = dfy.values.reshape(-1,)
xt = dfxtest.values
plt.scatter(x[:,2],x[:,3],c=y)
```

```python
def distance(x,q):
    return np.sqrt(sum((x-q)**2))
```

```python
def knn(x,y,q,k=25):
    vals = []
    for i in range(x.shape[0]):
        d = distance(x[i],q)
        vals.append((d,y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    data = np.unique(vals[:,1],return_counts = True)
    ind = data[1].argmax()
    result = data[0][ind]
    return int(result)
```

```python
def get_acc(x_tst,x,y):
    
    y_pred = []
    cnt=0
    for i in range(x_tst.shape[0]):
        p = knn(x,y,x_tst[i])
        if p==y[i]:
            cnt+=1
        
    #y_pred = np.array(y_pred)
    
    return  float(cnt)/x_tst.shape[0]
```

```python
query_point = x[3]

res = get_acc(xt,x,y)
print(res)
```
