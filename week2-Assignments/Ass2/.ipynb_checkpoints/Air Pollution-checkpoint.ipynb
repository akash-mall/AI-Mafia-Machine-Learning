```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dfx = pd.read_csv('./Train Data/Train.csv')
dftx = pd.read_csv('./Test Data/Test.csv')
x = dfx.values[:,:-1]
y = dfx.values[:,-1]
y = y.reshape((-1,1))
x1 = dftx.values

X = (x-x.mean())/x.std()
Y=y
X1 = (x1-x1.mean())/x1.std()
print(Y.shape)
```

```python
plt.scatter(X[:,0],Y,color='red')
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.score(X,Y)
```

```python
output = model.predict(X1)
print(output.shape)
print(X1.shape)
```

```python
plt.scatter(X[:,2],Y,label = 'data', color='red')
plt.plot(X1[:,2],output,color='blue',label = 'prediction')
plt.legend()
plt.show()
```
