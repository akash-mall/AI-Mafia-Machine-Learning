```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dfx = pd.read_csv('./Train Data/Train.csv')
dftx = pd.read_csv('./Test Data/Test.csv')
x = dfx.values[:,:-1]
y = dfx.values[:,-1]

x1 = dftx.values
X = (x-x.mean())/x.std()
Y=y
X1 = (x1-x1.mean())/x1.std()
print(X[0].shape[0])
```

```python
plt.plot(X,Y)
plt.show()
```

```python
#hypothesis

def hypothesis(X,theta,bias):
    hx = 0;
    for i in range(theta.shape[0]):
        hx += X[i]*theta[i]
    return hx+bias
           
```

```python
#cost/error function

def error(X,Y,theta,bias):
    m = X.shape[0]
    err = 0
    for i in range(m):
        hx = hypothesis(X[i],theta,bias)
        err += (hx-Y[i])**2
        
    return err/(2*m)
            
```

```python
#gradient

def gradient (X,Y,theta,bias):
    m = X.shape[0]
    n = X.shape[1]
    gradm = np.zeros((n))
    gradb = 0
    
    
    for i in range(m):
        hx = hypothesis(X[i],theta,bias)
        gradb += (hx-Y[i])/m
        for j in range(n):
            gradm[j] += ((hx-Y[i])*X[i][j])/m
            
    return (gradm,gradb)
```

```python
#gradient Descent

def gradientDescent(X,Y,learning_rate = 0.01):
    theta = np.ones(X[0].shape[0])
    bias = 1.5
    itr  = 0
    max_itr = 900
    m = X.shape[0]
    n = X.shape[1]
    error_list = []
    theta_list = []
    
    while (itr<=max_itr):
        gradm,gradb = gradient(X,Y,theta,bias)
        err = error(X,Y,theta,bias)
        error_list.append(err)
        theta_list.append(theta)
        for j in  range(n):            
            theta[j] = theta[j]-learning_rate*gradm[j]
        bias = bias - learning_rate*(gradb)
        itr+=1
                
    return theta,bias,error_list,theta_list
```

```python
final_theta,final_bias,err_list,thet_list = gradientDescent(X,Y)
print(final_theta,final_bias)
plt.plot(err_list)
plt.show()

```
